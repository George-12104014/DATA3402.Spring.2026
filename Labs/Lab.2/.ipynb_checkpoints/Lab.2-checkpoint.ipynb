*Exercise 1:* Write a function that creates an **n×n matrix** (a list of lists) representing the state of a tic‑tac‑toe game.

Use the integers:

- `0` = empty
- `1` = `"X"`
- `2` = `"O"`


````python
def create_board(n):
    """
    Creates an n x n tic-tac-toe board.

    0 = empty
    1 = X
    2 = O
    """
    if not isinstance(n, int) or n <= 0:
        print("Invalid board size.")
        return None

    board = []

    for i in range(n):
        row = []
        for j in range(n):
            row.append(0)
        board.append(row)

    return board
````

````python
board_3 = create_board(3)

print("Created board:")
print(board_3)

if board_3 == [[0,0,0],[0,0,0],[0,0,0]]:
    print("Exercise 1 working correctly.")
else:
    print("Something is wrong.")
````

*Exercise 2:* Write a function that takes two integers `n` and `m` and **draws** an `n` by `m` game board.

For example, the following is a 3×3 board:

```
   --- --- --- 
  |   |   |   | 
   --- --- ---  
  |   |   |   | 
   --- --- ---  
  |   |   |   | 
   --- --- --- 
```


````python
def draw_empty_board(n, m):
    if not isinstance(n, int) or not isinstance(m, int) or n <= 0 or m <= 0:
        print("Invalid board size.")
        return

    horizontal = " ---" * m
    vertical = "|" + ("   |" * m)

    for _ in range(n):
        print(horizontal)
        print(vertical)
    print(horizontal)
````

````python
draw_empty_board(3, 3)
````

*Exercise 3:* Modify Exercise 2 so that it takes a matrix in the format from Exercise 1 and draws a tic‑tac‑toe board with `"X"`s and `"O"`s.

````python
def draw_tictactoe(board):
    n = len(board)
    horizontal = " ---" * n

    for i in range(n):
        print(horizontal)
        row = "|"
        for j in range(n):
            if board[i][j] == 0:
                row += "   |"
            elif board[i][j] == 1:
                row += " X |"
            elif board[i][j] == 2:
                row += " O |"
            else:
                row += " ? |"
        print(row)
    print(horizontal)
````

````python
test_board = [[1,0,2],[0,2,0],[1,1,0]]
draw_tictactoe(test_board)
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
def winner(board):
    n = len(board)

    # rows
    for i in range(n):
        if board[i][0] != 0:
            win = True
            for j in range(n):
                if board[i][j] != board[i][0]:
                    win = False
            if win:
                return board[i][0]

    # cols
    for j in range(n):
        if board[0][j] != 0:
            win = True
            for i in range(n):
                if board[i][j] != board[0][j]:
                    win = False
            if win:
                return board[0][j]

    # diag \
    if board[0][0] != 0:
        win = True
        for i in range(n):
            if board[i][i] != board[0][0]:
                win = False
        if win:
            return board[0][0]

    # diag /
    if board[0][n-1] != 0:
        win = True
        for i in range(n):
            if board[i][n-1-i] != board[0][n-1]:
                win = False
        if win:
            return board[0][n-1]

    return 0
````

````python
A=[[2, 2, 0],
	[2, 1, 0],
	[2, 1, 1]]
print(winner(A))
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
def place_mark(board, player, row, col):
    n = len(board)

    if player not in (1, 2):
        return False
    if row < 0 or row >= n or col < 0 or col >= n:
        return False
    if board[row][col] != 0:
        return False

    board[row][col] = player
    return True
````

````python

b = create_board(3)
r1 = place_mark(b, 1, 0, 0)   # True
r2 = place_mark(b, 2, 0, 0)   # False (occupied)
r3 = place_mark(b, 2, 5, 5)   # False (out of bounds)
print(r1, r2, r3)
print(b)
if r1==True and r2==False and r3==False and b[0][0]==1:
    print("Ex5 PASS")
else:
    print("Ex5 FAIL")
````

*Exercise 6:* Modify Exercise 3 to show **row and column labels** so that players can specify locations like `"A2"` or `"C1"`.

````python
def draw_labeled_board(board):
    n = len(board)

    print("   ", end="")
    for i in range(n):
        print(str(i+1).rjust(2), end="  ")
    print()

    horizontal = "  " + " ---" * n

    for i in range(n):
        print(horizontal)
        row_label = chr(ord('A') + i)
        print(row_label, end="|")
        for j in range(n):
            v = board[i][j]
            if v == 0:
                print("   |", end="")
            elif v == 1:
                print(" X |", end="")
            else:
                print(" O |", end="")
        print()
    print(horizontal)
````

````python
b = create_board(3)
place_mark(b, 1, 0, 0)
place_mark(b, 2, 1, 1)
draw_labeled_board(b)
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
def convert_location(location, n):
    location = location.strip().upper()
    if len(location) < 2:
        return None

    row_char = location[0]
    if row_char < 'A' or row_char > chr(ord('A') + n - 1):
        return None

    num = location[1:]
    if not num.isdigit():
        return None

    row = ord(row_char) - ord('A')
    col = int(num) - 1
    if col < 0 or col >= n:
        return None

    return (row, col)
````

````python

print("A1 ->", convert_location("A1", 3))
print("c3 ->", convert_location("c3", 3))
print("D1 ->", convert_location("D1", 3))
if convert_location("A1",3)==(0,0) and convert_location("c3",3)==(2,2) and convert_location("D1",3) is None:
    print("Ex7 PASS")
else:
    print("Ex7 FAIL")
````

*Exercise 8:* Write a function that is called with a board and player number, takes input from the player using Python's `input()`, and modifies the board using your function from Exercise 7.

Keep asking for input until the player enters a valid location that results in a valid move.


````python
def player_move(board, player):
    n = len(board)
    while True:
        move = input("Enter move (e.g., A1): ")
        rc = convert_location(move, n)
        if rc is None:
            print("Invalid format. Try again.")
            continue
        r, c = rc
        if place_mark(board, player, r, c):
            break
        else:
            print("Spot taken or out of range. Try again.")
````

````python
b = create_board(3)
draw_labeled_board(b)

player_move(b, 1)   # এখানে A1 টাইপ move দেবে
draw_labeled_board(b)

player_move(b, 2)   # এখানে B2 টাইপ move দেবে
draw_labeled_board(b)
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
def is_full(board):
    for row in board:
        for cell in row:
            if cell == 0:
                return False
    return True
````

````python
b_full = [[1,2,1],[2,1,2],[2,1,2]]
print("Should be True:", is_full(b_full))

b_not_full = [[1,0],[2,1]]
print("Should be False:", is_full(b_not_full))
````

*Exercise 10:* Test that your game works for **5×5** tic‑tac‑toe.

````python
def play_game(n):
    board = create_board(n)
    player = 1

    while True:
        draw_labeled_board(board)
        player_move(board, player)

        w = winner(board)
        if w != 0:
            draw_labeled_board(board)
            print("Player", w, "wins!")
            return w

        if is_full(board):
            draw_labeled_board(board)
            print("Draw!")
            return 0

        player = 2 if player == 1 else 1
````

````python
play_game(3)
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
def available_moves(board):
    moves = []
    n = len(board)
    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                moves.append((i, j))
    return moves

````

````python
b = [[1,0],[2,0]]
print("Available moves should be [(0,1),(1,1)] or same positions:", available_moves(b))
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
def simple_computer_move(board, player):
    moves = available_moves(board)
    if len(moves) == 0:
        return False
    r, c = moves[0]
    place_mark(board, player, r, c)
    return True


````

````python
b = create_board(3)
ok = simple_computer_move(b, 2)
print("Move made:", ok)
draw_labeled_board(b)
if ok == True:
    print("Ex12 PASS")
else:
    print("Ex12 FAIL")
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
def computer_vs_computer(n):
    board = create_board(n)
    player = 1

    while True:
        moved = simple_computer_move(board, player)
        if not moved:
            print("Draw (no moves).")
            return 0

        w = winner(board)
        if w != 0:
            # draw_labeled_board(board)  # optional
            print("Player", w, "wins!")
            return w

        if is_full(board):
            print("Draw!")
            return 0

        player = 2 if player == 1 else 1

````

````python
result = computer_vs_computer(3)
print("Result:", result, "(0=draw, 1=X, 2=O)")
````
